```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

import requests
from bs4 import BeautifulSoup
import pandas as pd 
import json
```

```python
data = pd.read_csv("WATCHLIST.csv", encoding = "ISO-8859-1")
for i in range(0,data.count()[1]):
    movieTitle=data.loc[i][['Title']][0]
    movieYear=data.loc[i][['Year']][0]
    URL = 'https://www.werstreamt.es/filme/?q='+movieTitle+""
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find_all('li',itemprop='itemListElement')
    for job_elem in results:
        if None in job_elem:
            continue
        result = job_elem.find('meta',itemprop='dateCreated')
        if result==None:
            continue
        if (str(movieYear) in result["content"]):
            availableAt=""
            streams = json.loads(job_elem['data-avastatus'])
            if streams!=[]:
                for key, value in streams.items():
                    if value:
                        if key=="10":
                            availableAt +="Amazon" +", "
                        if key=="11":
                            availableAt +="Netflix" +", "
                        if key=="12":
                            availableAt +="Sky Go" +", "
            if availableAt!="" :
                print(color.BOLD+"("+str(movieYear)+") "+ movieTitle+color.END+" - "+availableAt, end='\n')
                break
```
